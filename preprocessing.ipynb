```python
import json

def replace_in_dict(data):
    if isinstance(data, dict):
        for key, value in data.items():
            if isinstance(value, (dict, list)):
                replace_in_dict(value)
            else:
                data[key] = str(value).replace('\r\n\r\n*', ' ').replace('\r\n\r\n', ' ').replace('\r\n\r', ' ').replace(
                    '\r\n', ' ').replace('\r', ' ').replace('\n', ' ').replace('*', ' ').replace('"', ' ')
                
    elif isinstance(data, list):
        for i, item in enumerate(data):
            if isinstance(item, (dict, list)):
                replace_in_dict(item)
            else:
                data[i] = str(item).replace('\r\n\r\n*', ' ').replace('\r\n\r\n', ' ').replace('\r\n\r', ' ').replace(
                    '\r\n', ' ').replace('\r', ' ').replace('\n', ' ').replace('*', ' ').replace('"', ' ')

    return data
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
def encoding(df,col_name, predefined_values):
    le = preprocessing.LabelEncoder()
    le.fit(list(predefined_values.keys()))
    df[col_name] = df[col_name].map(predefined_values)
    return df
```

```python
os.makedirs('encoded')
json_sentiment_path = 'lessen_columns/'
for full_name in os.listdir(json_sentiment_path):
    name = full_name.split(".")[0]
    type = name.split('_')[1]
    df = pd.read_csv(json_sentiment_path+full_name)
    if type== 'repos':
        print('\033[3m\033[1m\033[33m'+type)
        encoding_dict = {
            'has_issues':{True: 1, False: 0 },
            'archived':{True: 1, False: 0 },
            'has_projects':{True: 1, False: 0 },
            'has_wiki':{True: 1, False: 0 },
            'has_pages':{True: 1, False: 0 },
            'has_discussions' :{True: 1, False: 0 },
            'is_template':{True: 1, False: 0 },
        }
        for col in [ 'has_issues','archived','has_projects','has_wiki', 'has_pages', 'has_discussions' ,'is_template']:
            if col in df.columns:
                df = encoding(df, col, encoding_dict[col] )
        df.to_csv('encoded/'+full_name, index=False)


    if type== 'commits':
        print('\033[3m\033[1m\033[33m'+type)
        encoding_dict = {
            'author.type':{'User': 1, 'Organization': 1, 'nan':0, 'Bot' :0},
            'committer.type':{'User': 1, 'Organization': 1, 'nan':0, 'Bot' :0},
        
        }
        for col in ['author.type',  'committer.type',]:
            if col in df.columns:
                df = encoding(df, col, encoding_dict[col] )
        df.to_csv('encoded/'+full_name, index=False)


    if type== 'issues':
        print('\033[3m\033[1m\033[33m'+type)
        encoding_dict = {
            'user.type':{'User': 1, 'Organization': 1, 'nan':0, 'Bot' :0},
            'milestone.state':{'open': 1, 'closed': 0, 'nan':0},
        }
        for col in [ 'user.type','milestone.state']:
            if col in df.columns:
                df = encoding(df, col, encoding_dict[col] )
        df.to_csv('encoded/'+full_name, index=False)
        
```
